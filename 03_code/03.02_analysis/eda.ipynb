```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#pandas settings
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('../../02_data/02.02_processed/acs_brfss_21_22.csv')
```

```python
# Distribution of Diabetes Prevalence
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(df['brfss21_diabetes'], kde=True)
plt.title('Distribution of Diabetes Prevalence 2021')
plt.subplot(1, 2, 2)
sns.histplot(df['brfss22_diabetes'], kde=True)
plt.title('Distribution of Diabetes Prevalence 2022')
plt.tight_layout()
plt.show()
```

```python
# Calculate change in diabetes prevalence
df['diabetes_change_21_22'] = df['brfss22_diabetes'] - df['brfss21_diabetes']
sns.histplot(df['diabetes_change_21_22'], kde=True)
plt.title('Distribution of Change in Diabetes Prevalence (2022 - 2021)')
plt.show()
print(df['diabetes_change_21_22'].describe())
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['acs21_pct_ses_pov'], kde=True, stat='density', element='step', fill=True, color='blue', alpha=0.5, label='Poverty Rate 2021')
sns.histplot(df['acs22_pct_ses_pov'], kde=True, stat='density', element='step', fill=True, color='red', alpha=0.5, label='Poverty Rate 2022')
plt.title('Distribution of Poverty Rate: 2021 vs 2022')
plt.xlabel('Poverty Rate (%)')
plt.legend()
plt.show()

# obesity rate comparison
plt.figure(figsize=(8, 6))
sns.histplot(df['brfss21_obesity'], kde=True, stat='density', element='step', fill=True, color='blue', alpha=0.5, label='Obesity Rate 2021')
sns.histplot(df['brfss22_obesity'], kde=True, stat='density', element='step', fill=True, color='red', alpha=0.5, label='Obesity Rate 2022')
plt.title('Distribution of Obesity Rate: 2021 vs 2022')
plt.xlabel('Obesity Rate (%)')
plt.legend()
plt.show()

# broadband adoption comparison
plt.figure(figsize=(8, 6))
sns.histplot(df['acs21_pct_dg_bb_int'], kde=True, stat='density', element='step', fill=True, color='blue', alpha=0.5, label='Broadband Adoption Rate 2021')
sns.histplot(df['acs22_pct_dg_bb_int'], kde=True, stat='density', element='step', fill=True, color='red', alpha=0.5, label='Broadband Adoption Rate 2022')
plt.title('Distribution of Broadband Adoption Rate: 2021 vs 2022')
plt.xlabel('Broadband Adoption Rate (%)')
plt.legend()

# diabetes prevalence
plt.figure(figsize=(8, 6))
sns.histplot(df['brfss21_diabetes'], kde=True, stat='density', element='step', fill=True, color='blue', alpha=0.5, label='Diabetes Prevalence Rate 2021')
sns.histplot(df['brfss22_diabetes'], kde=True, stat='density', element='step', fill=True, color='red', alpha=0.5, label='Diabetes Prevalence Rate 2022')
plt.title('Distribution of Diabetes Prevalence Rate: 2021 vs 2022')
plt.xlabel('Diabetes Prevalence Rate (%)')
plt.legend()
plt.show()

```

```python

plt.figure(figsize=(8, 6))
df['acs_pct_pov_change'] = df['acs22_pct_ses_pov'] - df['acs21_pct_ses_pov']
sns.scatterplot(data=df, x='acs_pct_pov_change', y='diabetes_change_21_22', alpha=0.5)
plt.title('Diabetes Prevalence vs Poverty Rate (Change from 2021-2022)')
plt.xlabel('Poverty Rate (%)')
plt.ylabel('Diabetes Prevalence (%)')
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 6))
df['brfss_obesity_change'] = df['brfss22_obesity'] - df['brfss21_obesity']
sns.scatterplot(data=df, x='brfss_obesity_change', y='diabetes_change_21_22', alpha=0.5)
plt.title('Diabetes Prevalence vs Obesity Rate (Change from 2021-2022)')
plt.xlabel('Obesity Rate (%)')
plt.ylabel('Diabetes Prevalence (%)')
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 6))
df['acs_pct_bb_change'] = df['acs22_pct_dg_bb_int'] - df['acs21_pct_dg_bb_int']
sns.scatterplot(data=df, x='acs_pct_bb_change', y='diabetes_change_21_22', alpha=0.5)
plt.title('Diabetes Prevalence vs Broadband Adoption Rate (Change from 2021-2022)')
plt.xlabel('Broadband Adoption Rate (%)')
plt.ylabel('Diabetes Prevalence (%)')
plt.grid(True)
plt.show()

```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='brfss21_diabetes', y='brfss22_diabetes', alpha=0.5)
plt.plot([min(df['brfss21_diabetes'].min(), df['brfss22_diabetes'].min()),
        max(df['brfss21_diabetes'].max(), df['brfss22_diabetes'].max())],
        [min(df['brfss21_diabetes'].min(), df['brfss22_diabetes'].min()),
        max(df['brfss21_diabetes'].max(), df['brfss22_diabetes'].max())],
        color='black', linestyle=':') # Add a y=x line
plt.title('Diabetes Prevalence 2022 vs 2021')
plt.xlabel('Diabetes Prevalence 2021 (%)')
plt.ylabel('Diabetes Prevalence 2022 (%)')
plt.grid(True)
plt.show()

#broadband adoption
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='acs21_pct_dg_bb_int', y='acs22_pct_dg_bb_int', alpha=0.5)
plt.plot([min(df['acs21_pct_dg_bb_int'].min(), df['acs22_pct_dg_bb_int'].min()),
        max(df['acs21_pct_dg_bb_int'].max(), df['acs22_pct_dg_bb_int'].max())],
        [min(df['acs21_pct_dg_bb_int'].min(), df['acs22_pct_dg_bb_int'].min()),
        max(df['acs21_pct_dg_bb_int'].max(), df['acs22_pct_dg_bb_int'].max())],
        color='black', linestyle=':') # Add a y=x line
plt.title('Broadband Adoption Rate 2022 vs 2021')
plt.xlabel('Broadband Adoption Rate 2021 (%)')
plt.ylabel('Broadband Adoption Rate 2022 (%)')
plt.grid(True)
plt.show()
```
