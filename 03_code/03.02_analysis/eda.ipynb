```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#pandas settings
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('../../02_data/02.02_processed/acs_brfss_21_22.csv')
```

```python
# Distribution of Diabetes Prevalence
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(df['brfss21_diabetes'], kde=True)
plt.title('Distribution of Diabetes Prevalence 2021')
plt.subplot(1, 2, 2)
sns.histplot(df['brfss22_diabetes'], kde=True)
plt.title('Distribution of Diabetes Prevalence 2022')
plt.tight_layout()
plt.show()
```

```python
# Calculate change in diabetes prevalence
df['diabetes_change_21_22'] = df['brfss22_diabetes'] - df['brfss21_diabetes']
sns.histplot(df['diabetes_change_21_22'], kde=True)
plt.title('Distribution of Change in Diabetes Prevalence (2022 - 2021)')
plt.show()
print(df['diabetes_change_21_22'].describe())
```

```python
# Example: Distribution of Poverty Rate 2022
sns.histplot(df['acs22_pct_ses_pov'], kde=True)
plt.title('Distribution of Poverty Rate 2022')
plt.show()

# Example: Distribution of Obesity Rate 2022
sns.histplot(df['brfss22_obesity'], kde=True)
plt.title('Distribution of Obesity Rate 2022')
plt.show()

# Example: Distribution of Broadband Access 2022
sns.histplot(df['acs22_pct_dg_bb_int'], kde=True)
plt.title('Distribution of Broadband Access Rate 2022')
plt.show()
```

```python

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='acs22_pct_ses_pov', y='diabetes_change_21_22', alpha=0.5)
plt.title('Diabetes Prevalence vs Poverty Rate (2022)')
plt.xlabel('Poverty Rate (%)')
plt.ylabel('Diabetes Prevalence (%)')
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='brfss22_obesity', y='diabetes_change_21_22', alpha=0.5)
plt.title('Diabetes Prevalence vs Obesity Rate (2022)')
plt.xlabel('Obesity Rate (%)')
plt.ylabel('Diabetes Prevalence (%)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='brfss21_diabetes', y='brfss22_diabetes', alpha=0.5)
plt.plot([min(df['brfss21_diabetes'].min(), df['brfss22_diabetes'].min()),
        max(df['brfss21_diabetes'].max(), df['brfss22_diabetes'].max())],
        [min(df['brfss21_diabetes'].min(), df['brfss22_diabetes'].min()),
        max(df['brfss21_diabetes'].max(), df['brfss22_diabetes'].max())],
        color='red', linestyle='--') # Add a y=x line
plt.title('Diabetes Prevalence 2022 vs 2021')
plt.xlabel('Diabetes Prevalence 2021 (%)')
plt.ylabel('Diabetes Prevalence 2022 (%)')
plt.grid(True)
plt.show()
```
